<font size="3"> **References/Vocabs:** </font>

<font size="2"> A *__payload__* is basically the total amount of weight that a rocket can get up through our atmosphere and into space. So the likelihood that the payload number is more accurate than the exact weights of each module is high, because deciding the payload will likely affect each of the other design decisions.

We know that the *__Saturn V payload was 43,500 kg__*, and the weights of the modules varied from mission to mission. So, to determine the ratios that will allow us to make predictions about the Artemis missions, we can use:

- Saturn V payload
- Mission sample weight
- Mission module weight

We're calling the two modules *__crewed area__* in the DataFrame because those are the parts of the spacecraft where the crew can be, and likely where the samples would also reside.
</font>

#### Exercise 4.1

```python
import pandas as pd
rock_samples = pd.read_csv('data/rocksamples.csv')

rock_samples['Weight (g)'] = rock_samples['Weight (g)'].apply(lambda x : x * 0.001)
rock_samples.rename(columns = {'Weight (g)' : 'Weight (kg)'}, inplace = True)
rock_samples.head()

missions = pd.DataFrame()
missions['Mission'] = rock_samples['Mission'].unique()
missions.head()

sample_total_weight = rock_samples.groupby('Mission')['Weight (kg)'].sum()
missions = pd.merge(missions, sample_total_weight, on = 'Mission')
missions.rename(columns = {'Weight (kg)' : 'Sample Weight (kg)'}, inplace = True)
missions['Weight Difference (kg)'] = missions['Sample Weight (kg)'].diff()
missions['Weight Difference (kg)'] = missions['Weight Difference (kg)'].fillna(value = 0)

missions['Lunar Module (LM)'] = ['Eagle (LM-5)', 'Intrepid (LM-6)', 'Antares (LM-8)', 'Falcon (LM-10)', 'Orion (LM-11)', 'Challenger (LM-12)']
missions['LM Mass (kg)'] = [15103, 15235, 15264, 16430, 16445, 16456]
missions['LM Mass Difference'] = missions['LM Mass (kg)'].diff()
missions['LM Mass Difference'] = missions['LM Mass Difference'].fillna(value = 0)

missions['Command Module (CM)'] =['Columbia (CSM-107)', 'Yankee Clipper (CM-108)', 'Kitty Hawk (CM-110)', 'Endeavor (CM-112)', 'Casper (CM-113)', 'America (CM-114)']
missions['CM Mass (kg)'] = [5560, 5609, 5758, 5875, 5840, 5960]
missions['CM Mass Difference'] = missions['CM Mass (kg)'].diff()
missions['CM Mass Difference'] = missions['CM Mass Difference'].fillna(value = 0)

missions['Total Weight (kg)'] = missions['LM Mass (kg)'] + missions ['CM Mass (kg)']
missions['Total Difference (kg)'] = missions['LM Mass Difference'] + missions['CM Mass Difference']

# Sample-to-Weight Ratio
saturnVPayload = 43500 #(kg)
missions['Crewed Area: Payload'] = missions['Total Weight (kg)']/saturnVPayload
missions['Sample: Crewed Area'] = missions['Sample Weight (kg)']/missions['Total Weight (kg)']
missions['Sample: Payload'] = missions['Sample Weight (kg)']/saturnVPayload 

missions
```

```python
import pandas as pd
rock_samples = pd.read_csv('data/rocksamples.csv')

rock_samples['Weight (g)'] = rock_samples['Weight (g)'].apply(lambda x : x * 0.001)
rock_samples.rename(columns = {'Weight (g)' : 'Weight (kg)'}, inplace = True)
rock_samples.head()

rock_samples['Remaining (kg)'] = rock_samples['Weight (kg)'] * (rock_samples['Pristine (%)'] * 0.01)
rock_samples.describe()

low_samples = rock_samples.loc[(rock_samples['Weight (kg)'] >= 0.16) & (rock_samples['Pristine (%)'] <= 50)]
low_samples.head()
low_samples.Type.unique()

needed_samples = low_samples[low_samples['Type'].isin(['Basalt', 'Breccia'])]
needed_samples.groupby('Type')['Weight (kg)'].count()
needed_samples.groupby('Type')['Weight (kg)'].sum()

needed_samples = pd.concat([needed_samples, rock_samples[rock_samples['Type'] == 'Crustal']])

needed_samples_overview = pd.DataFrame()
needed_samples_overview['Type'] = needed_samples.Type.unique()

needed_samples_weights = needed_samples.groupby('Type')['Weight (kg)'].sum().reset_index()
needed_samples_overview = pd.merge(needed_samples_overview, needed_samples_weights, on = 'Type')
needed_samples_overview.rename(columns = {'Weight (kg)' : 'Total Weight (kg)'}, inplace = True)

needed_samples_ave_weights = needed_samples.groupby('Type')['Weight (kg)'].mean().reset_index()
needed_samples_overview = pd.merge(needed_samples_overview, needed_samples_ave_weights, on = 'Type')
needed_samples_overview.rename(columns = {'Weight (kg)' : 'Average Weight (kg)'}, inplace = True)

total_rock_count = rock_samples.groupby('Type')['ID'].count().reset_index()
needed_samples_overview = pd.merge(needed_samples_overview, total_rock_count, on = 'Type')
needed_samples_overview.rename(columns = {'ID' : 'Number of Samples'}, inplace = True)

total_rocks = needed_samples_overview['Number of Samples'].sum()
needed_samples_overview['Percentage of Rocks (%)'] = needed_samples_overview['Number of Samples']/total_rocks
needed_samples_overview['Percentage of Rocks (%)'] = needed_samples_overview['Percentage of Rocks (%)'].apply(lambda x : x * 100)

missions = pd.DataFrame()
missions['Mission'] = rock_samples['Mission'].unique()
missions.head()

sample_total_weight = rock_samples.groupby('Mission')['Weight (kg)'].sum()
missions = pd.merge(missions, sample_total_weight, on = 'Mission')
missions.rename(columns = {'Weight (kg)' : 'Sample Weight (kg)'}, inplace = True)
missions['Weight Difference (kg)'] = missions['Sample Weight (kg)'].diff()
missions['Weight Difference (kg)'] = missions ['Weight Difference (kg)'].fillna(value = 0)

missions['Lunar Module (LM)'] = ['Eagle (LM-5)', 'Intrepid (LM-6)', 'Antares (LM-8)', 'Falcon (LM-10)', 'Orion (LM-11)', 'Challenger (LM-12)']
missions['LM Mass (kg)'] = [15103, 15235, 15264, 16430, 16445, 16456]
missions['LM Mass Difference (kg)'] = missions['LM Mass (kg)'].diff()
missions['LM Mass Difference (kg)'] = missions['LM Mass Difference (kg)'].fillna(value = 0)

missions['Command Module (CM)'] =['Columbia (CSM-107)', 'Yankee Clipper (CM-108)', 'Kitty Hawk (CM-110)', 'Endeavor (CM-112)', 'Casper (CM-113)', 'America (CM-114)']
missions['CM Mass (kg)'] = [5560, 5609, 5758, 5875, 5840, 5960]
missions['CM Mass Difference (kg)'] = missions['CM Mass (kg)'].diff()
missions['CM Mass Difference (kg)'] = missions['CM Mass Difference (kg)'].fillna(value = 0)

missions['Total Weight (kg)'] = missions['LM Mass (kg)'] + missions['CM Mass (kg)']
missions['Total Difference (kg)'] = missions['LM Mass Difference (kg)'] + missions['CM Mass Difference (kg)']

# Sample-to-weight ratio for Apollo missions:
saturnVPayload = 43500 #(kg)
missions['Crewed Area: Payload'] = missions['Total Weight (kg)']/saturnVPayload
missions['Sample: Crewed Area'] = missions['Sample Weight (kg)']/missions['Total Weight (kg)']
missions['Sample: Payload'] = missions['Sample Weight (kg)']/saturnVPayload

missions

# Take the average of all those ratios across all the Apollo missions:
crewedarea_payload_average = missions['Crewed Area: Payload'].mean()
sample_crewedarea_average = missions['Sample: Crewed Area'].mean()
sample_payload_average = missions['Sample: Payload'].mean()

print(crewedarea_payload_average)
print(sample_crewedarea_average)
print(sample_payload_average)

artemis_crewedArea = 26520
artemis_missions = pd.DataFrame({'Mission': ['Artemis 1','Artemis 1b','Artemis 2'],
                                'Total Weight (kg)': [artemis_crewedArea,artemis_crewedArea,artemis_crewedArea],
                                'Payload (kg)': [26988, 37965, 42955]})

# Sample-to-weight ratio for (estimated) Artemis missions:
artemis_missions['Sample Weight from Total Weight (kg)'] = artemis_missions['Total Weight (kg)'] * sample_crewedarea_average
artemis_missions['Sample Weight from Payload (kg)'] = artemis_missions['Payload (kg)'] * sample_payload_average

artemis_missions['Estimated Sample Weight (kg)'] = (artemis_missions['Sample Weight from Total Weight (kg)'] + artemis_missions['Sample Weight from Payload (kg)'])/2

artemis_missions

artemis_ave_weight = artemis_missions['Estimated Sample Weight (kg)'].mean()
artemis_ave_weight

needed_samples_overview
```
